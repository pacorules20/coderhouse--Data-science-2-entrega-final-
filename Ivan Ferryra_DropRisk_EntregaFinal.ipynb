## 1- Introducción
El abandono universitario es un problema que afecta a miles de estudiantes y a las propias instituciones educativas. Hay un montón de factores que influyen en si un alumno sigue con sus estudios o decide dejarlos, desde cuestiones académicas hasta temas económicos y personales. Este dataset junta información clave sobre estudiantes de distintas carreras de grado, con la idea de analizar su desempeño y predecir si van a seguir o no en la universidad.



## 2- Descripción del problema

Hoy en día, detectar a tiempo qué estudiantes están en riesgo de abandonar la facultad no es nada fácil. Hay muchas variables en juego y no siempre hay herramientas que permitan hacer un análisis preciso. Esto genera problemas para:

Universidades: porque se complica armar estrategias para evitar la deserción.

Estudiantes: que muchas veces no reciben la ayuda necesaria para seguir adelante.

Gobiernos y organismos educativos: que no cuentan con información concreta para diseñar políticas de retención más efectivas.

La falta de modelos predictivos basados en datos dificulta la toma de decisiones y el desarrollo de mejores estrategias para mejorar la tasa de egreso.

## 3- Objetivo y Alcance

### Objetivo Principal:

Crear un modelo de clasificación que pueda predecir si un estudiante va a abandonar, continuar o terminar la carrera en base a su rendimiento académico y su situación socioeconómica.

### Objetivos Específicos:

  * Identificar qué factores tienen más peso en la deserción universitaria.

  * Diseñar y evaluar modelos de machine learning para predecir la continuidad de los estudiantes.

  * Brindar una herramienta útil para que las universidades puedan tomar medidas preventivas.

### Alcance:

  * Se analizarán datos de estudiantes de distintas carreras de grado.

  * Se pondrá el foco en el rendimiento durante el primer y segundo semestre.

  * Se planteará el problema como una clasificación en tres categorías: abandono, continuidad y graduación.

  * Se aplicarán técnicas de machine learning para mejorar la precisión del modelo.

  * Se van a utilizar estrategias para equilibrar los datos y evitar sesgos.

El modelo que salga de este estudio va a ser clave para ayudar a las universidades a tomar mejores decisiones y a los estudiantes a recibir el apoyo que necesitan para terminar sus estudios.

ATRIBUTO → DESCRIPCIÓN

1. Estado civil → Situación marital del estudiante (soltero, casado, divorciado, etc.).

2. Modo de aplicación → Forma en la que el estudiante presentó su solicitud de ingreso (por ejemplo, examen de admisión, traslado, etc.).

3. Orden de aplicación → Número de veces que el estudiante ha solicitado admisión en la institución. (entre 0 y 9)

4. Curso → Programa o carrera en la que el estudiante está matriculado (lista completa en la fuente).

5. Tipo de asistencia (diurna/nocturna) → Indica si el estudiante asiste a clases en el turno diurno o nocturno.

6. Calificación previa → Última titulación o nivel educativo alcanzado antes de ingresar a la institución(lista completa en la fuente).

7. Nota de la calificación previa → Puntaje obtenido en la última titulación o nivel educativo previo. (entre 0 y 200)

8. Nacionalidad → País de origen del estudiante(lista completa en la fuente).

9. Calificación de la madre → Nivel educativo alcanzado por la madre del estudiante (lista completa en la fuente).

10. Calificación del padre → Nivel educativo alcanzado por el padre del estudiante (lista completa en la fuente).

11. Ocupación de la madre → Profesión o actividad laboral de la madre del estudiante (lista completa en la fuente).

12. Ocupación del padre → Profesión o actividad laboral del padre del estudiante (lista completa en la fuente).

13. Nota de admisión → Puntaje obtenido en el proceso de admisión de la institución.(entre 0 y 200)

14. Desplazado → Indica si el estudiante ha sido desplazado por conflictos o razones socioeconómicas. (si o no)

15. Necesidades educativas especiales → Indica si el estudiante requiere apoyo educativo especial. (si o no)

16. Deudor → Indica si el estudiante tiene deudas pendientes con la institución.(si o no)

17. Cuotas de matrícula al día → Indica si el estudiante ha pagado todas sus cuotas de matrícula. (si o no)

18. Género → Identidad de género del estudiante. (hombre o mujer)

19. Becado → Indica si el estudiante recibe una beca para financiar sus estudios. (si o no)

20. Edad al matricularse → Edad del estudiante al momento de ingresar a la institución.

21. Internacional → Indica si el estudiante es extranjero. (si o no)

22. Unidades curriculares 1er semestre (convalidadas) → Materias del primer semestre que fueron validadas por estudios previos.

23. Unidades curriculares 1er semestre (inscritas) → Cantidad de materias en las que el estudiante se inscribió en el primer semestre.

24. Unidades curriculares 1er semestre (evaluadas) → Cantidad de materias del primer semestre en las que el estudiante presentó evaluaciones.

25. Unidades curriculares 1er semestre (aprobadas) → Cantidad de materias del primer semestre que el estudiante aprobó.

26. Nota de unidades curriculares 1er semestre → Promedio de calificaciones obtenidas en las materias del primer semestre.

27. Unidades curriculares 1er semestre (sin evaluar) → Materias inscritas en el primer semestre en las que el estudiante no presentó evaluaciones.

28. Unidades curriculares 2do semestre (convalidadas) → Materias del segundo semestre que fueron validadas por estudios previos.

29. Unidades curriculares 2do semestre (inscritas) → Cantidad de materias en las que el estudiante se inscribió en el segundo semestre.

30. Unidades curriculares 2do semestre (evaluadas) → Cantidad de materias del segundo semestre en las que el estudiante presentó evaluaciones.

31. Unidades curriculares 2do semestre (aprobadas) → Cantidad de materias del segundo semestre que el estudiante aprobó.

32. Nota de unidades curriculares 2do semestre → Promedio de calificaciones obtenidas en las materias del segundo semestre.

33. Unidades curriculares 2do semestre (sin evaluar) → Materias inscritas en el segundo semestre en las que el estudiante no presentó evaluaciones.

34. Tasa de desempleo → Porcentaje de personas desempleadas en el país o región en un período determinado.

35. Tasa de inflación → Cambio en el nivel de precios de bienes y servicios en un período determinado.

36. PIB (Producto Interno Bruto) → Valor total de bienes y servicios producidos en un país en un período determinado.

37. Objetivo (Target) → Variable de salida del modelo, que clasifica a los estudiantes en una de tres categorías al final de la duración normal del curso: abandono (dropout), inscrito (enrolled) o graduado (graduate).

```python
def download_and_read_csv_from_url(url):
    csv_path = "./data.csv"

    # Descargar el archivo directamente desde la URL
    response = requests.get(url)
    response.raise_for_status()  # Lanza una excepción si hay un error en la descarga

    # Guardar el contenido en un archivo local
    with open(csv_path, "wb") as file:
        file.write(response.content)

    # Leer el CSV descargado
    df = pd.read_csv(csv_path)
    return df
```

## Graficos Exploratorios

### Hipotesis 1 - ¿Es el rendimiento académico en el segundo semestre un predictor más fuerte del logro del objetivo que el rendimiento en el primer semestre?

El gráfico muestra la relación entre el rendimiento académico en los primeros dos semestres y el objetivo final del estudiante. A partir de estos gráficos, podemos extraer las siguientes conclusiones:

  - Bajo rendimiento en estudiantes que abandonan: Presentan menores notas promedio, alta dispersión y muchos casos con calificación 0.

  - Rendimiento intermedio en estudiantes aún cursando: Sus notas se ubican entre las de los graduados y los que abandonan, lo que indica un posible punto de inflexión.

### Hipotesis 3 - ¿Son las variables demográficas son significativos del logro del objetivo académico?

```python
# Hipótesis 3: Las variables demográficas y socioeconómicas no son predictores significativos del logro del objetivo académico.

# Variables demográficas
demograficas = ['Estado civil', 'Género', 'Nacionalidad']

plt.figure(figsize=(18, 10))
for i, col in enumerate(demograficas):
    # Calcular el porcentaje de cada valor en la columna actual
    total_estudiantes = len(df)
    porcentajes = df[col].value_counts(normalize=True) * 100

    # Filtrar los valores que representan menos del 1%
    valores_filtrados = porcentajes[porcentajes >= 1].index
    df_filtrado = df[df[col].isin(valores_filtrados)]

    # Crear el gráfico de barras agrupadas con datos filtrados
    plt.subplot(2, 3, i + 1)
    sns.countplot(x=col, hue='Objetivo', data=df_filtrado )
    plt.title(f'{col} vs Objetivo (Valores >= 1%)')
    plt.xlabel(col)
    plt.ylabel('Cantidad de estudiantes')
    plt.legend(title='Objetivo', labels=['abandono', 'graduado', 'cursando'])

plt.tight_layout()
plt.show()
```

Los gráficos muestran la relación entre variables sociodemográficas (estado civil, género y nacionalidad) y el objetivo final del estudiante. A partir de ellos, podemos extraer las siguientes conclusiones:

  - Predominio de estudiantes solteros/as (estado civil = 1): La mayoría de los estudiantes, sin importar el objetivo, pertenece a esta categoría. Sin embargo, dentro de este grupo, hay más graduados que abandonos o cursantes.

  - Género y abandono: El abandono es más frecuente entre estudiantes del grupo de hombres, mientras que los del grupo 1 tienen una mayor proporción de graduados.

  - Nacionalidad dominante (Portuguesa): La mayoría de los estudiantes pertenecen a una sola nacionalidad, que concentra la mayor cantidad de graduados y también de abandonos, aunque en menor medida. Otras nacionalidades no superan el 1% y fueron excluidas del gráfico.

### Hipotesis 4 - Los estudiantes que no rinden evaluaciones en el primer semestre tienen una menor probabilidad de alcanzar su objetivo académico

```python
# Recordemos como estaba compuesto nuestro preprocesador
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer()),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, lista_numeric_features),
        ('cat', categorical_transformer, lista_categorical_features)])
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

# 1. Binarizar las etiquetas (por ejemplo, de [0,1,2] a [[1,0,0],[0,1,0],[0,0,1]])
y_test_binarized = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_binarized.shape[1]

# 2. Obtener probabilidades (OvR)
y_score = grid_cv.predict_proba(X_test)

# 3. Curvas ROC por clase
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# 4. Graficar todas las curvas ROC
plt.figure()
colors = ['blue', 'red', 'green', 'orange', 'purple']
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i % len(colors)],
             label=f'Clase {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curvas ROC por Clase (Multiclase)')
plt.legend(loc="lower right")
plt.grid()
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize
import numpy as np

y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
y_score = grid_cv.predict_proba(X_test)

print("AUC macro:", roc_auc_score(y_test_bin, y_score, average="macro", multi_class="ovr"))
print("AUC micro:", roc_auc_score(y_test_bin, y_score, average="micro", multi_class="ovr"))
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Predicciones
y_train_pred = grid_cv.predict(X_train)
y_test_pred = grid_cv.predict(X_test)

# Evaluación en entrenamiento
print("🔹 Métricas sobre el conjunto de ENTRENAMIENTO:")
print(classification_report(y_train, y_train_pred))
print(f"Accuracy en entrenamiento: {accuracy_score(y_train, y_train_pred):.2f}")
print("-" * 60)

# Evaluación en test
print("🔹 Métricas sobre el conjunto de TEST:")
print(classification_report(y_test, y_test_pred))
print(f"Accuracy en test: {accuracy_score(y_test, y_test_pred):.2f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Predicción
y_pred = grid_cv.predict(X_test)

# Matrices
cm = confusion_matrix(y_test, y_pred)
cm_norm = confusion_matrix(y_test, y_pred, normalize='true')

# Figuras lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Matriz no normalizada
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues", ax=axes[0], colorbar=False)
axes[0].set_title("Matriz de Confusión")
axes[0].grid(False)
for spine in axes[0].spines.values():
    spine.set_visible(False)

# Matriz normalizada
disp_norm = ConfusionMatrixDisplay(confusion_matrix=cm_norm)
disp_norm.plot(cmap="Blues", ax=axes[1], colorbar=True)
axes[1].set_title("Matriz de Confusión Normalizada")
axes[1].grid(False)
for spine in axes[1].spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```
